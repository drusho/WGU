```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
```

```python
ibm = pd.read_csv('Telco_customer_churn.csv')
ibm.head()
```

```python
ibm.info()
```

```python
ibm.duplicated(keep='last')
```

```python
ibm.isnull().sum()
```

```python
ibm.columns
```

```python
ibm.drop(['Count', 'Country', 'Lat Long', 'Total Charges', 'Churn Value', "Multiple Lines", 'Partner',
       'Churn Score', 'CLTV', 'Churn Reason'], axis=1, inplace=True)
```

```python
ibm.shape
```

```python
ibm.info()
```

```python
ibm= ibm.rename(columns = {"customerID": "Customer_id", "Churn Label": "Churn", "Senior Citizen": "Age", "Dependents": "Children",
                    "Tenure Months": "Tenure", "Monthly Charges": "MonthlyCharge", "Phone Service": "Phone",
                          "Internet Service": "InternetService", "Online Security": "OnlineSecurity", "Online Backup": "OnlineBackup",
                          "Device Protection": "DeviceProtection", "Tech Support": "TechSupport", "Streaming TV": "StreamingTV", "Streaming Movies": "StreamingMovies",
                          "Paperless Billing": "PaperlessBilling", "Payment Method": "PaymentMethod"})
```

```python
ibm.head()
```

```python
ibm.describe()
```

```python
ibm['Contract'].unique()
```

```python
ibm['Contract'] = ibm['Contract'].replace("Two year","Two Year")
```

```python
ibm['PaymentMethod'].unique()
```

```python
ibm['PaymentMethod'] = ibm['PaymentMethod'].replace('Electronic check', 'Electronic Check')
ibm['PaymentMethod'] = ibm['PaymentMethod'].replace('Mailed check', 'Mailed Check')
ibm['PaymentMethod'] = ibm['PaymentMethod'].replace('Bank transfer (automatic)', 'Bank Transfer(automatic)')
ibm['PaymentMethod'] = ibm['PaymentMethod'].replace('Credit card (automatic)', 'Credit Card (automatic)')
```

```python
ibm['InternetService'].unique()
```

```python
ibm['InternetService'] = ibm['InternetService'].replace('Fiber optic', 'Fiber Optic')
ibm['InternetService'] = ibm['InternetService'].replace('No', 'None')
```

```python
ibm['Gender'].unique()
```

```python
ibm['State'].unique()
```

```python
ibm['State'] = ibm['State'].replace('California', 'CA')
```

```python
ibm['OnlineSecurity']=ibm['OnlineSecurity'].astype('category')
ibm['OnlineSecurity']=ibm['OnlineSecurity'].cat.codes
```

```python
ibm['OnlineBackup']=ibm['OnlineBackup'].astype('category')
ibm['OnlineBackup']=ibm['OnlineBackup'].cat.codes
```

```python
ibm['DeviceProtection']=ibm['DeviceProtection'].astype('category')
ibm['DeviceProtection']=ibm['DeviceProtection'].cat.codes
```

```python
ibm['TechSupport']=ibm['TechSupport'].astype('category')
ibm['TechSupport']=ibm['TechSupport'].cat.codes
```

```python
ibm['StreamingTV']=ibm['StreamingTV'].astype('category')
ibm['StreamingTV']=ibm['StreamingTV'].cat.codes
```

```python
ibm['StreamingMovies']=ibm['StreamingMovies'].astype('category')
ibm['StreamingMovies']=ibm['StreamingMovies'].cat.codes
```

```python
ibm.Age = np.where(ibm.Age == "Yes", "Senior Citizen",
                               np.where(ibm.Age == "No", "Adult", "Adult"))
```
